# Hanlding the mission values  

### categorical features  

```python
cat_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1
                   and dataset[feature].dtypes == 'O']

print('there are {} categorical nan features'.format(len(cat_with_nan)))

# getting the percentage of nan in features 

for feature in cat_with_nan:
       print( feature , dataset[feature].isnull().mean()*100)
```

#### first step toward hadnling the categorical nan feature is to replace the nan with some new label ( say 'missing' )  

```python
# creating a function which will convert the nan to 'Missing'

def replace_cat_feature(feature_list, dataset):
    
    data = dataset.copy()
    
    data[feature_list] = data[feature_list].fillna('missing')
    
    return data 
```

```python
dataset = replace_cat_feature(cat_with_nan, dataset)
```

```python
dataset[cat_with_nan].isna().sum()
```

```python
# let's see how many numeric features have nan 

numeric_with_nan = [feature for feature in dataset.columns if dataset[feature].isna().sum() > 1 and dataset[feature].dtypes != 'O']

print('there are {} numeric feature that contains nan '.format(len(numeric_with_nan)))

dataset[numeric_with_nan].isna().mean()*100
```

#### whenever we have lot of outliers in the data set , then we should fill the nan with median or mode  

```python
# replacing the numerical nan feature with median 

for feature in numeric_with_nan:
    
    median = dataset[feature].median()
    
    # creating a new feature , just to capture the nan, this help us to understand the whether there was a nan or not  
    
    dataset[feature + '_nan'] = np.where(dataset[feature].isna(), 1, 0)
    dataset[feature] = dataset[feature].fillna(median)
    
dataset[numeric_with_nan].isna().sum()
```

## handling temporal variables 

```python
# this is very important stuff to handle , beacuse the price of the house will be decided the no. of years 
# we will transform the each year feature with no. of years 
# it means we will calculate that, how many years it is old , how many years does the garage is old and how many ears ago it was remodifies 

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']: 
    
    dataset[feature] = dataset['YrSold'] - dataset[feature]  # it will cature the no. of years 
```

```python
dataset.head(4)
```

### log normal distribution 
#### we have seen in exploratory data anlayis that , the data is skwed and we need to transform it into log normal distribution 

```python
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    dataset[feature] = np.log(dataset[feature])
```

# handling rare categorical features
####  it says that , if in a categorical feature if some labels are less than 1% of the total observation , than we can skip that particular label 

```python
categorical_feature  = dataset.select_dtypes('O').columns 
categorical_feature
```

```python
dataset.groupby('MSZoning')['SalePrice'].count()/len(dataset)    # it means that , how many times does the each label came 
                                                    #, while calculating the sale price
```

```python
dataset['MSZoning'].value_counts()/len(dataset)
```

```python
for feature in categorical_feature:
        per = dataset[feature].value_counts()/len(dataset)
        per_greater_than_1 = per[per>0.01].keys()
        dataset[feature] = np.where(dataset[feature].isin(per_greater_than_1), dataset[feature], 'rare')        
```

```python
dataset.head()
```

# label encoding  

```python
# performing label encoding 

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for feature in categorical_feature:
    dataset[feature] = le.fit_transform(dataset[feature])
```

```python
dataset.head()
```

# feature scaling 
#### we will do feature scaling beacuse in this data set we have so many features which might have been recorded in different different units 
#### hence to get the better accuracy, it is important to perform feature scaling 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['Id', 'SalePrice']]
```

```python
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
# getting the entire data frame

data = pd.concat([dataset[['Id', 'SalePrice']], pd.DataFrame(scaler.transform(dataset[feature_scale]), 
                                                      columns = feature_scale)], axis = 1)
```
